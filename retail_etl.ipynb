```python
#data loading
import pandas as pd

#different kaggles dataset workwith latin-1
df_raw = pd.read_csv("superstore_raw.csv", encoding="latin-1")
df_raw.head()



#first row
df_raw.head()
```

```python
#cleaning
#standardize or  fix column names
df_raw.columns=df_raw.columns.str.strip()                     #remove spaces
df_raw.columns=df_raw.columns.str.replace("","_")             #replace spaces with _
df_raw.columns=df_raw.columns.str.lower()                      #lowercases
```

```python
print(df_raw.columns.tolist())    #show column names
```

```python
import numpy as np
#handel null values
df_raw=df_raw.dropna(subset=["Order ID"])    #drop rows missing values in OrderID
df_raw.fillna({"Country":0, "City":0, "Quantity":0},inplace=True)    #fill rows with given values 


#drive unit price


if "unit_price" not in df_raw.columns:
    df_raw["unit_price"] = np.where(
        df_raw["Quantity"] > 0,
        df_raw["Sales"] / df_raw["Quantity"],
        np.nan
    )

#create total price column
df_raw["total_price"]=df_raw["Quantity"]*df_raw["unit_price"]

#save clean data
df_raw.to_csv("superstore_clean.csv", encoding="latin-1",index=False)

df_raw.head()
    
```

```python
query = 'SELECT "Category", SUM("total_price") AS total_sales FROM superstore GROUP BY "Category";'
pd.read_sql(query, engine)
```

```python
#Top 5 selling products

query = """
SELECT "Product Name", SUM("total_price") AS total_sales
FROM superstore
GROUP BY "Product Name"
ORDER BY total_sales DESC
LIMIT 5;
"""
pd.read_sql(query, engine)

```

```python
#Total revenue by region
query = """
SELECT "Region", SUM("total_price") AS total_revenue
FROM superstorejl
GROUP BY "Region"
ORDER BY total_revenue DESC;
"""
pd.read_sql(query, engine)
```

```python
#Monthly revenue trend

query = """
SELECT 
    DATE_TRUNC('month', TO_DATE("Order Date", 'DD/MM/YYYY')) AS month,
    SUM("total_price") AS monthly_revenue
FROM superstore
GROUP BY month
ORDER BY month;
"""
pd.read_sql(query, engine)
```
